```python
import math

sin_rom = [int(2**10 * math.sin((x/4096)* 2 * math.pi)) for x in range(4096)]
for a in range(len(sin_rom)):
    print("sine[{index}]={value};".format(index=a, value=sin_rom[a]))
```

```python
import pandas as pd
pd.DataFrame(sin_rom).plot()
```
